```python
# gotta import the things I think I'll need
import numpy as np
import pandas as pd
import nilearn as nil
import os
import re
from sklearn.preprocessing import normalize
from nilearn.input_data import NiftiMasker 
# Change dir so I'm working from the top level
os.chdir('../..')
output_dir='MVPA/data'
behave_input='EV_files_m12_trial_by_trial/'
mr_input='cardtask/'
#participants 120,... are missing 1 or more runs
```

```python
#Let's make the image files
#list of subjects
part_list=os.listdir(mr_input)
trial_img_list=[]
normalize=NiftiMasker(standardize=True)
for sub in part_list:
    sub_out=output_dir+'/sub-'+sub
    if not os.path.exists(sub_out):
        os.makedirs(sub_out)
    print(sub)
    sub_files=[]
    dirs=os.listdir(mr_input+sub)
    
    #make list of behavioral files
    beh_files=os.listdir(behave_input+sub)
    beh_files=[s for s in beh_files if 'trial_info.txt' in s]
    beh_files.sort(key=lambda x:re.search('_r(.*)',x).group(0))
    beh_files=[behave_input+sub+'/'+s for s in beh_files]
    print beh_files
    r1=pd.read_csv(beh_files[0],delimiter='\t',header=None)
    r1['type']=re.search('/%s/(.*)_r1'%(sub),beh_files[0]).group(1)
    r1['run']=1
    r2=pd.read_csv(beh_files[1],delimiter='\t',header=None)
    r2['type']=re.search('/%s/(.*)_r2'%(sub),beh_files[1]).group(1)
    r2['run']=2
    r3=pd.read_csv(beh_files[2],delimiter='\t',header=None)
    r3['type']=re.search('/%s/(.*)_r3'%(sub),beh_files[2]).group(1)
    r3['run']=3
    r4=pd.read_csv(beh_files[3],delimiter='\t',header=None)
    r4['type']=re.search('/%s/(.*)_r4'%(sub),beh_files[3]).group(1)
    r4['run']=4

    dfs=[r1,r2,r3,r4]
    df=pd.concat(dfs)
    df.to_csv(sub_out+'/sub-'+sub+'_labels.csv',sep='\t')
    for run in [1,2,3,4]:
        print run
        #getting the image data into a list of files
        stats_dir=[s for s in dirs if 'Stats1_run%s'%(run) in s]
        dest=mr_input+sub+'/'+stats_dir[0]+'/stats'
        print dest
        fnames=[f for f in os.listdir(dest) if os.path.isfile(os.path.join(dest,f))]
        full_path=[]
        for f in fnames:
            full_path.append(os.path.abspath(dest+'/'+f))
        full_path.sort()
        sub_files.append(full_path)
    
    #print sub_files
    #run1=nil.image.load_img(sub_files[0])
    #run1=normalize.fit_transform(run1)
    #run1=normalize.inverse_transform(run1)
    
    #run2=nil.image.load_img(sub_files[1])
    #run2=normalize.fit_transform(run2)
    #run2=normalize.inverse_transform(run2)

    #run3=nil.image.load_img(sub_files[2])
    #run3=normalize.fit_transform(run3)
    #run3=normalize.inverse_transform(run3)

    #run4=nil.image.load_img(sub_files[3])
    #run4=normalize.fit_transform(run4)
    #run4=normalize.inverse_transform(run4)

    
    sub_files=np.concatenate(sub_files)
    
    np.savetxt(sub_out+'/sub-'+sub+'_mrfiles.csv',
               sub_files,
               delimiter=',',
               fmt='%s')
    
    #full_image=nil.image.concat_imgs([run1,run2,run3,run4])
    full_image=nil.image.load_img(sub_files)
    
    func_img_name=sub_out+'/sub-'+sub+'_func.nii.gz'
    print(func_img_name)
    full_image.to_filename(func_img_name)
```
